```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 100)

y1 = 3*np.cos(2*x)-2*np.sin(3*x)
y2 = -7*x+3
ax = plt.subplot()
ax.plot(x, y2, 'b')
ax.plot(x, y1, 'r--')
ax.set(title='y=3cos(2x)-2sin(3x) et y=-7x+3', ylabel='Axe Y', xlabel='Axe X')
plt.legend(['Droite', 'Courbe'])
plt.grid(True, which='both')
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
plt.xlim(-np.pi, np.pi)
plt.ylim(-20, 25)
plt.show()
```
